Task 4

```python
# Generate x values from 1 to 100
x = np.arange(1, 101)  


y1 = 2 * x + 1
y2 = 3 * x**2
y3 = np.cos(x)

plt.figure(figsize=(12, 4))  # Set the figure size

# Plot for y = 2x + 1
plt.subplot(131)  # Subplot 1
plt.plot(x, y1)
plt.title('y = 2x + 1')

# Plot for y = 3x^2
plt.subplot(132)  # Subplot 2
plt.plot(x, y2)
plt.title('y = 3x^2')

# Plot for y = cos(x)
plt.subplot(133)  # Subplot 3
plt.plot(x, y3)
plt.title('y = cos(x)')

# Show the plots
plt.tight_layout()  # Adjust the layout
plt.show()
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('lab4_dataset1.csv')

# Extract columns x1 and x2
x1 = df['x1']
x2 = df['x2']

# Create a scatter plot
plt.scatter(x1, x2, c='blue', marker='+', label='Scatter Plot')

# Set axis labels and plot title
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Scatter Plot of x1 vs. x2')

# Show the plot
plt.show()
```

PART C



```python
# Load the cleaned dataset
df_cleaned = pd.read_csv('lab5_task3.csv')
```

```python
x1_column = 'value' 
x2_column = 'population'  
```

```python
# Scatter plot settings
markers = {0: 'o', 1: 's'}  # 0: Red circles, 1: Blue squares

# Create a scatter plot based on 'inland' column
for label, marker in markers.items():
    filtered_df = df_cleaned[df_cleaned['inland'] == label]
    plt.scatter(filtered_df[x1_column], filtered_df[x2_column], marker=marker, label=f'inland={label}')

# Set axis labels and plot title
plt.xlabel(x1_column)
plt.ylabel(x2_column)
plt.title(f'Scatter Plot of {x1_column} vs. {x2_column} by Inland')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
